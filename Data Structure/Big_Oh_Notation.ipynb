you were assigned to decide which algorithms you should use between these two

````
# Algorithm 1

def Add_up_1(N): 
  i = 0
  for n in range(N):
    i += n 
  return i
````

````
# Algorithm 2

def Add_up_2(N):
  return N*(N-1)/2
````

````
Add_up_1(100)
````

````
Add_up_2(100)
````

- With your bare eye, you may say Algorithm 2 is faster, because it took less code lines. The answer is True but not because of less code lines:

- Algorithm 2 is faster: No matter N what, algorithm 2 always takes 3 simple operations regardless of the size of N: 1 multiplication, 1 subtraction, 1 division


```
def Add_up_2(N):
  return N*(N-1)/2      #1 multiplication, 1 subtraction, 1 division
```

- Algorithm 1 is slower: It takes one assignment (i=0) 


```
def Add_up_1(N):       
  i = 0                # 1 assignment
  for n in range(N):   # N assignments (for n) 
    i += n             # N addition 
  return i

```






In order to compare, you might need to use Big Oh notation

> Algorithm 1 : O(N) <br>
> Algorithm 2 : O(1) <br><br>
since Algorithm 2 always take **constant** time which is 3, no matter how large N was, 
<br>since Algorithm 1 take **linear** time which is N, the speed depends on N size <br><br>


> you might ask why the Big oh Notation is 
> Algorithm 1 : O(2N+1) <br>
Algorithm 2 : O(3) <br><br>
The answer is because Big Oh doesn't focus on details, but big picture!! (as N increase)

#### 2) Linear time ((O(n))

Always found in <font color='red'>**Linear search**</font>

````
def linear_search(data, value):
    for index in range(len(data)):
        if value == data[index]:
            return index
    raise ValueError('Value not found in the list')
    
if __name__ == '__main__':
    data = [1, 2, 9, 8, 3, 4, 7, 6, 5]
    print(linear_search(data, 7))
````

Some Application Example of O(n^2) : <font color='red'>Bubble sort</font>

````
def bubble_sort(data):
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(data)-1):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                swapped = True
    
if __name__ == '__main__':
    data = [9, 1, 7, 6, 2, 8, 5, 3, 4, 0]
    bubble_sort(data)
    print(data)
````

#### 5) <u>Quasilinear Time — O(n log n)</u>

An algorithm is said to have a quasilinear time complexity when each operation in the input data have a logarithm time complexity. It is commonly seen in sorting algorithms (e.g. **<font color='red'>mergesort</font>**, timsort, heapsort).
For example: for each value in the data1 (O(n)) use the binary search (O(log n)) to search the same value in data2.

````
def merge_sort(data):
    if len(data) <= 1:
        return
    
    mid = len(data) // 2
    left_data = data[:mid]
    right_data = data[mid:]
    
    merge_sort(left_data)
    merge_sort(right_data)
    
    left_index = 0
    right_index = 0
    data_index = 0
    
    while left_index < len(left_data) and right_index < len(right_data):
        if left_data[left_index] < right_data[right_index]:
            data[data_index] = left_data[left_index]
            left_index += 1
        else:
            data[data_index] = right_data[right_index]
            right_index += 1
        data_index += 1
    
    if left_index < len(left_data):
        del data[data_index:]
        data += left_data[left_index:]
    elif right_index < len(right_data):
        del data[data_index:]
        data += right_data[right_index:]
    
if __name__ == '__main__':
    data = [9, 1, 7, 6, 2, 8, 5, 3, 4, 0]
    merge_sort(data)
    print(data)
````

#### 6) Exponential Time — O(2^n)



- Often seen in **<font color='red'>Recursion Algorithm</font>** (such as Fibonacci)

````
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
````
